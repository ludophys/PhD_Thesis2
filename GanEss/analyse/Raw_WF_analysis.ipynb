```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter

# --- Fonction Landau (approchée) ---
def landau(x, A, mu, c):
    x_shifted = (x - mu) / c
    return A * np.exp(-0.5 * (x_shifted + np.exp(-x_shifted)))

# --- Générer des données bruitées ---
np.random.seed(42)
x = np.linspace(-2, 6, 500)
true_params = [1.0, 1.5, 0.4]  # A, mu, c
y = landau(x, *true_params) + 0.02 * np.random.normal(size=len(x))

# --- Lissage des données pour une meilleure estimation initiale ---
y_smooth = savgol_filter(y, window_length=31, polyorder=3)

# --- Estimation automatique améliorée ---
def estimate_initial_params(x, y):
    A0 = np.max(y)
    mu0 = x[np.argmax(y)]
    half_max = A0 / 2
    indices = np.where(y >= half_max)[0]
    if len(indices) > 1:
        c0 = (x[indices[-1]] - x[indices[0]]) / 2
    else:
        c0 = 0.5
    return [A0, mu0, c0]

initial_guess = estimate_initial_params(x, y_smooth)

# --- Fit avec restriction de plage autour du pic ---
fit_window = 2.5  # fenêtre autour du maximum pour le fit
x_fit_mask = (x > initial_guess[1] - fit_window) & (x < initial_guess[1] + fit_window)
x_fit = x[x_fit_mask]
y_fit = y[x_fit_mask]

params, _ = curve_fit(landau, x_fit, y_fit, p0=initial_guess)

# --- Courbe ajustée complète ---
y_model = landau(x, *params)

# --- Tracé ---
plt.figure(figsize=(9, 5))
plt.plot(x, y, label='Données bruitées', alpha=0.5)
plt.plot(x, y_smooth, label='Lissé (savgol)', linestyle='--', alpha=0.6)
plt.plot(x, y_model, 'r', label='Fit Landau')
plt.axvline(params[1], color='green', linestyle='--', label='μ ajusté')
plt.xlabel("x")
plt.ylabel("Amplitude")
plt.title("Fit amélioré d'une distribution de Landau")
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()

# --- Affichage des résultats ---
print("\n--- Paramètres ajustés ---")
param_names = ["Amplitude (A)", "Centre (μ)", "Largeur (c)"]
for name, val in zip(param_names, params):
    print(f"{name} = {val:.4f}")
```
